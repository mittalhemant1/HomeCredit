```python
application_train = pd.read_csv("application_train.csv")
```

```python
application_train.describe()
```

```python
data_summary = pd.DataFrame(application_train.dtypes)
data_summary.columns = ['dtypes']
data_summary['variable'] = data_summary.index
data_summary['non_missing'] = application_train.count()
data_summary['unique'] = application_train.nunique()
data_summary['missing_pct'] = 1- data_summary['non_missing']/application_train.shape[0]
data_summary['unique_pct'] = 1- data_summary['unique']/application_train.shape[0]
data_summary.reset_index(drop=True, inplace=True)
```

```python
category = [x[0] for x in data_summary.loc[data_summary['unique'] <60,['variable']].values.tolist()]
```

```python
univariate = application_train[category].apply(pd.Series.value_counts).unstack().dropna().to_frame() 
univariate.columns = ['count']
univariate['percent'] = univariate['count']/application_train.shape[0]
univariate.reset_index(inplace=True)
univariate.rename(columns={'level_0':'variable','level_1':'levels'}, inplace=True)
```

```python
View(univariate)
```

```python
bivariate = application_train[category].apply(lambda x : application_train.groupby([x,'TARGET']).size()).stack().dropna().to_frame()
```

```python
bivariate.index.names = ['levels','TARGET','variable']
bivariate.columns = ['counts']
```

```python
bivariate_ordered = bivariate.pivot_table(index=['variable','levels'],columns=['TARGET'],values=['counts'],aggfunc = np.sum)
bivariate_ordered.columns = ['target_0','target_1']
bivariate_ordered.fillna(0, inplace=True)
bivariate_ordered.reset_index(inplace=True)
```

```python
uni_bi = pd.merge(bivariate_ordered,univariate,on=['variable','levels'])
```

```python
events = uni_bi.loc[(uni_bi['variable']=='TARGET') & (uni_bi['levels']==1),'target_1'].iloc[0]
non_events = uni_bi.loc[(uni_bi['variable']=='TARGET')& (uni_bi['levels']==0) ,'target_0'].iloc[0]
print(events,non_events)
```

```python
uni_bi['hit_rate'] = uni_bi['target_1']/uni_bi['count']
uni_bi['pct_events'] = uni_bi['target_1']/events
uni_bi['pct_non_events'] = uni_bi['target_0']/non_events
```

```python
uni_bi['woe'] = np.log(uni_bi['pct_events']/uni_bi['pct_non_events'])
uni_bi.replace([np.inf, -np.inf],[1,-1],inplace=True)
uni_bi['IV'] = uni_bi['woe'] * (uni_bi['pct_events']-uni_bi['pct_non_events'])
```

```python
IV = uni_bi.groupby(['variable'])['IV'].sum().to_frame().replace([np.inf, -np.inf],1).reset_index()
uni_bi.replace([np.inf, -np.inf],1,inplace=True)
IV.columns=['variable', 'IV_variable']
```

```python
uni_bi_woe = pd.merge(uni_bi,IV , on = ['variable'])
```
